--------------------  n_components should be ===> 8   --------------------------

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=8)
pca.fit(X_scaled)
```

```python
explained_variance= pca.explained_variance_
print('Explained Variance:\n',explained_variance)
```

```python
components=pca.components_
print('Components:\n',components)
```

```python
X_final_transformed=pca.transform(X_scaled)
```

---------------------------------------Which Stocks are similar in performance----------------------------------------------

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,31):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X_final_transformed)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(10,8))    
plt.plot(range(1,31),wcss)
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
k_means=KMeans(n_clusters=10,random_state=0,init='k-means++')
k_means.fit(X_final_transformed)
y_kmeans=kmeans.fit_predict(X_final_transformed)
labels=k_means.labels_
```
